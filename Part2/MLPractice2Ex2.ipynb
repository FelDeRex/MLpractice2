```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Olympics2018.xlsx", sheet_name='Olympics')

df1 = df[(df['Country'] == 'Russia') & (df['Medal'] == 'GOLD')]
fig, ax = plt.subplots()
ax.scatter(df1['Age'], df1['Event'], c = 'g')
ax.set_facecolor('black')     
fig.set_figwidth(16)
fig.set_figheight(14)
plt.savefig("MLpractice2Graph1.pdf")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Olympics2018.xlsx", sheet_name='Olympics')

df = df.groupby('Year').Age.mean().dropna()
a = plt.plot(df.index, df.values)
plt.savefig("MLpractice2Graph2.pdf")
plt.show()
```

```python
import pandas as pd

df = pd.read_excel("Olympics2018.xlsx", sheet_name='Olympics')

df = df[df['Season'] == 'Summer']
df = df.drop(['City'], axis=1)
df['Birth date'] = df['Year'] - df['Age']
print(df)

df.to_excel("NewOlymics2018.xlsx")
```
